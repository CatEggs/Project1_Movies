```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import locale # to format currency as USD
locale.setlocale( locale.LC_ALL, '' )

# Import API key
from api_keys import api_key

# Output File (CSV)
output_data_file = "tmbddata.csv"

## Generate Movies Database from TMBD API

r = requests.get('https://api.themoviedb.org/3/discover/movie?api_key=' +  api_key)
highest_revenue_ever = r.json()
r.json()

```

```python
columns = ['film', release_date', vote_average, popularity]
```

```python
#Create dataframe from JSON

columns = ['film', 'revenue', 'budget', 'release_date']
df = pd.DataFrame(columns=columns)
for film in highest_revenue_films_ever:
    film_revenue = requests.get('https://api.themoviedb.org/3/movie/'+ str(film['id']) +'?api_key='+ api_key+'&language=en-US')
    film_revenue = film_revenue.json()
    df.loc[len(df)]=[film['title'],film_revenue['revenue'], (film_revenue['budget'] * -1), film_revenue['release_date']]
    
    
    
```
